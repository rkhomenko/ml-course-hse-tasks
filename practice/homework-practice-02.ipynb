```python
def pairplot(df, target):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * 4, nrow * 4))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature)
        if i % ncol == 0:
            plt.ylabel('target')
```

Постройте график зависимости $R^2$ для обучающей и тестовой (на кросс-валидации) выборок в зависимости от значения гиперпараметра. Для Lasso также постройте график зависимости количества ненулевых весов.

__Lasso__

```python
alphas = np.power(10.0, np.arange(-6.0, 7.0, 1.0))
scores_train = []
scores_test = []
nonzero_weights = []
for alpha in alphas:
    model = Lasso(alpha)
    model.fit(x_train, y_train)
    scores_train.append(np.mean(cross_val_score(model, x_train, y_train, cv=5)))
    scores_test.append(np.mean(cross_val_score(model, x_test, y_test, cv=5)))
    nonzero_weights.append(model.coef_[model.coef_ > 0.1].shape[0])
```

Зачастую признаки в сырых данных имеют разный масштаб. Попробуйте применить масштабирование к данным, сравните качество. Заметно ли изменение?

Pipeline, StandardScaler, MinMaxScaler

```python
models = [(Pipeline([('scaler', StandardScaler()), ('reg', Lasso())]), 'Lasso StandardScaler'),
          (Pipeline([('scaler', MinMaxScaler()), ('reg', Lasso())]), 'Lasso MinMaxScaler'),
          (Pipeline([('scaler', StandardScaler()), ('reg', Ridge())]), 'Ridge Standard'),
          (Pipeline([('scaler', MinMaxScaler()), ('reg', Ridge())]), 'Ridge MinMaxScaler')]

def calc_score(model, alphas, train_ds, test_ds, need_nonzero_weights=False):
    x_train, y_train = train_ds
    x_test, y_test = test_ds
    
    cv = GridSearchCV(estimator=model,
                      param_grid={
                          'reg__alpha': alphas
                      })
    cv.fit(x_train, y_train)
    
    print(f'Best params: {cv.best_params_}')
    
    scores_train = []
    scores_test = []
    nonzero_weights = []
    for alpha in alphas:
        model.set_params(reg__alpha=alpha)
        model.fit(x_train, y_train)
        scores_train.append(np.mean(cross_val_score(model, x_train, y_train, cv=5)))
        scores_test.append(np.mean(cross_val_score(model, x_test, y_test, cv=5)))
        nonzero_weights.append(model[-1].coef_[model[-1].coef_ > 0.1].shape[0])
    
    if need_nonzero_weights:
        return scores_train, scores_test, nonzero_weights
    
    return scores_train, scores_test, None
```

```python
def plot_results(label, alphas, scores_train, scores_test, nonzero_weights=None):
    if nonzero_weights is not None:
        f, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
    else:
        f, ax1 = plt.subplots(1, 1, figsize=(13, 5))

    ax1.plot(np.log10(alphas), scores_train, label='train')
    ax1.plot(np.log10(alphas), scores_test, label='test')
    ax1.set_title(f"Зависимость R^2 от alpha ({label})")
    ax1.set_xlabel("log10(Alpha)")
    ax1.set_ylabel("R^2")
    ax1.grid()
    ax1.legend()

    if nonzero_weights is not None:
        ax2.plot(np.log10(alphas), nonzero_weights, label='weights')
        ax2.set_title(f"Зависимость числа ненулевых весов от alpha ({label})")
        ax2.set_xlabel("log10(Alpha)")
        ax2.set_ylabel("Count")
        ax2.grid()
        ax2.legend()
    
    plt.plot()
```

```python
for model, label in models:
    if 'Lasso' in label:
        need_weights = True
    else:
        need_weights = False
    score_train, score_test, nonzero_weights = calc_score(model,
                                                          alphas,
                                                          (x_train, y_train),
                                                          (x_test, y_test),
                                                          need_weights)
    plot_results(label, alphas, scores_train, scores_test, nonzero_weights)
```

```python
ridge_std = create_model(Ridge(), StandardScaler())
ridge_std.fit(x_train_trf, y_train)
ridge_std.score(x_test_trf, y_test), ridge_std.best_params_
```
